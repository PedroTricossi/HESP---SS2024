```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data with pandas
df_nb_list = pd.read_csv("nb_list_time.csv", sep=",")  # Use ';' as your delimiter
df_cut_off = pd.read_csv("cut_off_radious_time.csv", sep=",")  # Use ';' as your delimiter
df_no_opt = pd.read_csv("no_opt.csv", sep=",")  # Use ';' as your delimiter

df_cut_off.head()
```

```python
num_particles = df_nb_list['num_particles']
nb_list_time = df_nb_list[' time']
cut_off_time = df_cut_off[' time']
no_opt_time = df_no_opt[' time']
```

```python
# Create the plot
plt.figure(figsize=(24, 18))

# Plot data for each test type
plt.semilogx(num_particles, nb_list_time, marker='o', linestyle='-', color='b', label='Neighbour List', base=2)
plt.semilogx(num_particles, cut_off_time, marker='s', linestyle='--', color='g', label='Cut off radious', base=2)
plt.semilogx(num_particles, no_opt_time, marker='^', linestyle='-.', color='r', label='No optimization', base=2)

plt.title("Time vs Number of Particles")
plt.xlabel("number of particles")
plt.ylabel("time (s)")
plt.grid(True)
plt.legend()

# Annotate data points
# for i, (x, y) in enumerate(zip(buffer_sizes, base_bandwidth)):
#     plt.annotate(f"Base ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')
# for i, (x, y) in enumerate(zip(buffer_sizes, openmp_bandwidth)):
#     plt.annotate(f"OpenMP ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')
# for i, (x, y) in enumerate(zip(buffer_sizes, cuda_bandwidth)):
#     if y is not None:
#         plt.annotate(f"CUDA ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')

# Show the plot
plt.savefig("time_vs_num_particles.png")
```
