```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data with pandas
df = pd.read_csv("results.csv", sep=";")  # Use ';' as your delimiter
base_data = df[df['test_type'] == 'base']
openmp_data = df[df['test_type'] == 'openmp']
cuda_data = df[df['test_type'] == 'cuda']

df.head()
```

```python
buffer_sizes = base_data["buffer_size"]
base_bandwidth = base_data["bandwidth"]
openmp_bandwidth = openmp_data["bandwidth"]
cuda_bandwidth = cuda_data["bandwidth"]
```

```python

# Create the plot
plt.figure(figsize=(24, 18))

# Plot data for each test type
plt.semilogx(buffer_sizes, base_bandwidth, marker='o', linestyle='-', color='b', label='Base', basex=2)
plt.semilogx(buffer_sizes, openmp_bandwidth, marker='s', linestyle='--', color='g', label='OpenMP', basex=2)
plt.semilogx(buffer_sizes, cuda_bandwidth, marker='^', linestyle='-.', color='r', label='CUDA', basex=2)

plt.title("Bandwidth vs. Buffer Size")
plt.xlabel("Buffer Size")
plt.ylabel("Bandwidth (GB/s)")
plt.grid(True)
plt.legend()

# Annotate data points
# for i, (x, y) in enumerate(zip(buffer_sizes, base_bandwidth)):
#     plt.annotate(f"Base ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')
# for i, (x, y) in enumerate(zip(buffer_sizes, openmp_bandwidth)):
#     plt.annotate(f"OpenMP ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')
# for i, (x, y) in enumerate(zip(buffer_sizes, cuda_bandwidth)):
#     if y is not None:
#         plt.annotate(f"CUDA ({x}, {y:.2f})", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')

# Show the plot
plt.show()
```
